## Gradient Descent
*  Effiecient code using Vectorization


```python
def hypo(X,theta):
    return np.dot(X,theta)


def error(X,y,theta):
    e=0.0
    m=X.shape[0]
    y_=hypo(X,theta)
    e=np.sum((y-y_)**2)
    return e/m

def gradient(X,y,theta):
    y_=hypo(X,theta)
    grad=np.dot(X.T,(y_-y))
    m=X.shape[0]
    
    
    return grad/m

def g_d(X,y,lr=0.1,epoch=300):
    n=X.shape[1]
    theta=np.zeros((n,))
    error_list=[]
    for i in range(epoch):
        e=error(X,y,theta)
        e=error_list.append(e)
        grad=gradient(X,y,theta)
        theta=theta-lr*grad
        
        
    return theta,error_list
```

```python
ft,error_list=g_d(X,y)

```

```python
plt.plot(error_list)
plt.show()
```

### Calculating Score of the prediction made by the algorithm

```python
def r2_score(y,j):
    num=np.sum((y-j)**2)
    dnom=np.sum((y-y.mean())**2)
    score=(1-(num/dnom))
    return score*100
```
